```python
# product table creation and data insertion
import mysql.connector
from mysql.connector import Error
import pandas as pd


try:
    # 1. Rename DataFrame columns to match SQL schema
    df_products = df_products.rename(columns={
        'Product Name': 'Product_Name',
        'Unit Cost USD': 'Unit_Cost_USD',
        'Unit Price USD': 'Unit_Price_USD'
    })

    # 2. Convert numeric columns and handle invalid values
    df_products['Unit_Cost_USD'] = pd.to_numeric(df_products['Unit_Cost_USD'], errors='coerce')
    df_products['Unit_Price_USD'] = pd.to_numeric(df_products['Unit_Price_USD'], errors='coerce')

    # 3. Drop rows where the price column is invalid (NaN) so insertion succeeds
    df_products = df_products.dropna(subset=['Unit_Price_USD'])

    # 4. Establish database connection
    connection = mysql.connector.connect(
         host = "gateway01.ap-northeast-1.prod.aws.tidbcloud.com",
        port = 4000,
        user = "3a4M49L7Ack2pZS.root",
        password = "ZYuwgIk8h7AFcEaE",
        database = "dataspark"
    )
        

       

    if connection.is_connected():
        cursor = connection.cursor()

        # 5. Create the Products table (if it doesn’t already exist)
        create_table_sql = '''
            CREATE TABLE IF NOT EXISTS Products (
                ProductKey INT,
                Product_Name VARCHAR(255),
                Brand VARCHAR(255),
                Color VARCHAR(255),
                Category VARCHAR(255),
                Unit_Cost_USD FLOAT,
                Unit_Price_USD FLOAT,
                Subcategory VARCHAR(255),
                CategoryKey VARCHAR(255)
            )
        '''
        cursor.execute(create_table_sql)
        print("Products table created successfully")

        # 6. Prepare the INSERT query
        insert_sql = '''
            INSERT INTO Products (
                ProductKey,
                Product_Name,
                Brand,
                Color,
                Category,
                Unit_Cost_USD,
                Unit_Price_USD,
                Subcategory,
                CategoryKey
            ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        '''

        # 7. Convert DataFrame to list of tuples, in the correct column order
        data_to_insert = list(
            df_products[
                ['ProductKey', 'Product_Name', 'Brand', 'Color', 'Category',
                 'Unit_Cost_USD', 'Unit_Price_USD', 'Subcategory', 'CategoryKey']
            ].itertuples(index=False, name=None)
        )

        # 8. Execute batch insertion and commit the transaction
        cursor.executemany(insert_sql, data_to_insert)
        connection.commit()
        print(f"{cursor.rowcount} records inserted successfully")

    else:
        print("Failed to connect to the database")

except Error as e:
    print(f"Error: {e}")

finally:
    # 9. Clean up resources (cursor + connection)
    if 'cursor' in locals() and cursor:
        cursor.close()
    if connection and connection.is_connected():
        connection.close()
        print("Connection closed")
```

```python
# sales table creation and data insertion
import mysql.connector
from mysql.connector import Error
import re

try:
    connection = mysql.connector.connect(
      host = "gateway01.ap-northeast-1.prod.aws.tidbcloud.com",
      port = 4000,
      user = "3a4M49L7Ack2pZS.root",
      password = "ZYuwgIk8h7AFcEaE",
      database = "dataspark"
)

    if connection.is_connected():
        cursor = connection.cursor()
        print("✅ Connected to MySQL")

        # 1️⃣ Create table with 9 columns (including StoreKey)
        create_table_sql = '''
            CREATE TABLE IF NOT EXISTS Sales (
                Order_Number INT,
                Line_Item INT,
                Order_Date VARCHAR(255),
                Delivery_Date VARCHAR(255),
                CustomerKey INT,
                StoreKey INT,
                ProductKey INT,
                Quantity INT,
                Currency_Code VARCHAR(255)
            )
        '''
        cursor.execute(create_table_sql)
        print("✅ Sales table (9 columns) ready")

        # 2️⃣ Clean and normalize columns automatically
        print("\nOriginal columns in df_sales:", df_sales.columns.tolist())

        def clean_column(col):
            col = col.strip()
            col = col.replace(" ", "_")
            col = re.sub(r"[^A-Za-z0-9_]", "", col)
            return col

        df_sales.columns = [clean_column(c) for c in df_sales.columns]
        print("🧹 Cleaned columns:", df_sales.columns.tolist())

        # 3️⃣ Expected 9 columns
        expected_cols = [
            'Order_Number', 'Line_Item', 'Order_Date', 'Delivery_Date',
            'CustomerKey', 'StoreKey', 'ProductKey', 'Quantity', 'Currency_Code'
        ]

        # 4️⃣ Auto-match similar names
        mapping = {}
        for col in df_sales.columns:
            for target in expected_cols:
                if col.lower().replace("_", "") == target.lower().replace("_", ""):
                    mapping[col] = target
                    break

        df_sales = df_sales.rename(columns=mapping)
        missing = [col for col in expected_cols if col not in df_sales.columns]
        if missing:
            raise KeyError(f"Missing expected columns in DataFrame: {missing}")

        df_sales = df_sales[expected_cols]
        print("✅ Final column order:", df_sales.columns.tolist())

        # 5️⃣ Insert into MySQL
        insert_sql = '''
            INSERT INTO Sales (
                Order_Number, Line_Item, Order_Date, Delivery_Date,
                CustomerKey, StoreKey, ProductKey, Quantity, Currency_Code
            ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        '''

        data_to_insert = [tuple(row) for row in df_sales.to_numpy()]
        print(f"\n📦 Ready to insert {len(data_to_insert)} records...")

        if len(data_to_insert) > 0:
            cursor.executemany(insert_sql, data_to_insert)
            connection.commit()
            print(f"✅ {cursor.rowcount} Sales records inserted successfully")
        else:
            print("⚠️ No records found in df_sales to insert")

        # 6️⃣ Verify total rows
        cursor.execute("SELECT COUNT(*) FROM Sales")
        total_rows = cursor.fetchone()[0]
        print("📊 Total rows in Sales table:", total_rows)

except Error as e:
    print(f"⚠️ Database Error: {e}")

except KeyError as e:
    print(f"⚠️ Column Error: {e}")

finally:
    if 'cursor' in locals() and cursor:
        cursor.close()
    if connection and connection.is_connected():
        connection.close()
        print("🔒 Connection closed")
```
